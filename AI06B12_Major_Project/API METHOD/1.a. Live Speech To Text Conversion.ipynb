### Importing Modules

```python
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
```

### Initialize the instance

```python
r = sr.Recognizer()
```

### Recognizing Live speech using Microphone

```python
with sr.Microphone() as source:
    print("Kindly speak through the microphone:\n")
    r.adjust_for_ambient_noise(source, duration=0.5)     #  Eliminating noise  
    audio = r.listen(source, phrase_time_limit=5)        #  Listening to Audio
    
    try:
        print("Converting your speech to text....\n")
        text = r.recognize_google(audio)                 #  Recognizing Audio using the Google Speech API
        
        print("Your speech converted to ENGLISH text is as follows:\n {}".format(text))    #  Printing the Text generated
        
        print("\nPossible Transcriptions:")                                        #  Printing all the possibilities
        dic = r.recognize_google(audio, show_all=True)
        print("{}".format(dic))
        
    except:
        print("Sorry couldn't recognize your voice")
```

### Recognizing speech using .wav file

```python
audio = 'audio.wav'

with sr.AudioFile(audio) as source:
    r.adjust_for_ambient_noise(source)        #  Eliminating noise
    audio_data= r.listen(source)                  #  Listening to Audio
    
    try:
        print("Converting your audio file to text....\n")
        
        print("Audio file converted to text is as follows:\n " + r.recognize_google(audio_data) )   #  Printing the Text generated
        print("\nPossible Transcriptions:")                                                    #  Printing all the possibilities
        dic = r.recognize_google(audio_data, show_all=True)
        print("{}".format(dic))
        
    except:
        print("Sorry couldn't recognize your voice")
```
